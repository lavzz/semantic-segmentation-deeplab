```python
"""
check the directory called 'models'. You should see .h5 weights files for xception and mobilenet backends 
uploaded...for ease of usage, we are going to copy the mobilenet version to the root directory and rename it to model_weights 
"""

import shutil 
from shutil import copyfile
src = 'models/deeplabv3_mobilenetv2_tf_dim_ordering_tf_kernels.h5'
dst = 'model_weights.h5'
copyfile(src, dst)
```

```python
"""
at this point we have everything needed to run inference. as a first step, lets define a model called deeplab_model and load our weights 
The backbone/weights we have chosen is for mobilenet and not xception . Mobilenet is geared towards smaller embedded chips found in 
mobile phones and IoT sensors . You can try the same steps with xception as well
"""

deeplab_model = Deeplabv3(input_shape = (512, 512,3), classes = 21, weights = 'pascal_voc', OS=8)

deeplab_model.load_weights('model_weights.h5')
```

```python
"""
we are going to run inference on some common coco images found in the imgs folder. 

"""
import imageio
image_example = imageio.imread('imgs/image3.jpg') ##change to other images found in teh folder like 'image1', 'image2' etc.
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 

plt.imshow(image_example)
```

```python
from PIL import Image

##pre-process to fit the deeplab training image size 

trained_image_width = 512
mean_subtraction_value = 127.5  
image = np.array(Image.open('imgs/image3.jpg'))

w, h, _ = image.shape
ratio = float(trained_image_width) / np.max([w, h])
resized_image = np.array(Image.fromarray(image.astype('uint8')).resize((int(ratio * h), int(ratio * w))))


resized_image = (resized_image/mean_subtraction_value)-1.
pad_x = int(trained_image_width - resized_image.shape[0])
pad_y = int(trained_image_width - resized_image.shape[1])
resized_image = np.pad(resized_image, ((0, pad_x), (0, pad_y), (0,0)),mode = 'constant')


```

```python
##add an extra dimension to be able to feed it into the neural network for inference 
resized_image = np.reshape(resized_image, [1, resized_image.shape[0], resized_image.shape[1], resized_image.shape[2]])
resized_image.shape
```

```python
labels = deeplab_model.predict(resized_image)
```

```python
labels.shape ##check that the shape of the prediced label is the same across dimensions 1 and 2 as input image and 21 in dimension #3 which is hte number of classes 
```

```python
label = np.argmax(labels.squeeze(), -1) ##argmax to project class information onto a single dimension 
```
